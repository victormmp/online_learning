# Toy Dataset - Análise

```python
import pandas as pd
import numpy as np
from collections import Counter
import csv
```

Primeiramente iremos carregar os dados contidos nos arquivos de texto. As amostras estão salvas em um arquivo diferente, `X_hyperplane.txt`, do que os labels, `y_hyperplane.txt`.

```python
with open('X_hyperplane.txt') as file:
    reader = csv.reader(file)
    x_data = []
    columns = None
    for row in reader:
        x_data.append(row)
        if columns is None:
            columns = [f'x{i}' for i in range(len(row))]
    x = pd.DataFrame(x_data, dtype=np.float, columns=columns)
        
with open('y_hyperplane.txt') as file:
    reader = csv.reader(file)
    y_data = []
    for row in reader:
        y_data += row
    y = pd.Series(y_data, dtype=int)
```

Cada amostra contém 11 features,

```python
x.head()
```

E a distribuição de cada uma das classes no conjunto de dados carregado,

```python
Counter(y)
```

Dividindo as amostras em conjuntos separados para cada classe, vemos as estatísticas básicas para cada uma das features.

```python
class_1 = x[y==1]
class_2 = x[y==-1]
```

```python
class_1.describe()
```

```python
class_2.describe()
```

Vemos que, para as duas classes, a variância é alta, sendo quase 50% do valor da média, para todas as features. Isto pode ser um indicativo de que existem diferentes distribuições de dados em cada feature, por algum *dataset shift*.
